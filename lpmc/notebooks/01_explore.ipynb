# 01 - Exploracion del LPMC

Notebook espejo de `01_explore.py` y del preprocesado del profe. Sirve para abrir el CSV bruto y dejar guardados resmenes de columnas en `artifacts/`.

```python
from pathlib import Path
import pandas as pd
import io, json

ROOT = Path.cwd()
RAW_PATH = ROOT / 'data' / 'raw' / 'LPMC_dataset.csv'
ARTIFACTS = ROOT / 'artifacts'
ARTIFACTS.mkdir(exist_ok=True)
print(f'Leyendo dataset bruto desde: {RAW_PATH}')
```

```python
df = pd.read_csv(RAW_PATH)
print('Forma del dataset:', df.shape)
df.head()
```

```python
buf = io.StringIO()
df.info(buf=buf)
print(buf.getvalue())
```

```python
summary_rows = []
for col in df.columns:
    s = df[col]
    n_unique = s.nunique(dropna=True)
    dtype = str(s.dtype)
    if n_unique <= 10:
        sample_vals = sorted(map(str, s.dropna().unique()))[:10]
    else:
        sample_vals = sorted(map(str, s.dropna().unique()))[:5]
    summary_rows.append({
        'column': col,
        'dtype': dtype,
        'non_null': int(s.notna().sum()),
        'n_unique': int(n_unique),
        'sample_values': ', '.join(sample_vals),
    })

summary_df = pd.DataFrame(summary_rows).sort_values('column')
summary_csv_path = ARTIFACTS / 'lpmc_column_summary.csv'
summary_json_path = ARTIFACTS / 'lpmc_column_summary.json'
summary_df.to_csv(summary_csv_path, index=False)
summary_csv_path
```
